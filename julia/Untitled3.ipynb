```julia
using DifferentialEquations

l  = 10
b = 0.02
a = 0.2
function neuron(du,u,p,t)
    du[1] = 0.04*u[1]^2 + 5*u[1]+140-u[2]+l
    du[2] = a*(b*u[1]-u[2])
end

u0 = [0.0;0.0;]
tspan = (0.0,100.0)
prob = SecondOrderODEProblem(neuron,du0,u0,tspan)
sol = solve(prob)
```

```julia
using Plots
plot(sol[1,:],sol[2,:])
```
