もっと華麗なコード（黒木さんのコード）

```julia
#using Seaborn

n = 10
N = 1000000

X = 3 .+ randn(N)                  # Normal(3,1)
Y = vec(mean(3 .+ randn(N, n),2)) # mean of size-n samples for Normal(3,1)
Z = 3 .+ randn(N)/√n               # Normal(3,1/√n)

kdeplot(X, label="Normal(3,1)")
kdeplot(Y, label="mean of size-$n samples for Normal(3,1)", ls="--")
kdeplot(Z, label="Normal(3, 1/√$n)", ls="-.")
grid(ls = ":")
xlim(-1,11)
legend(fontsize=9)
```
