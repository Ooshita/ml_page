```julia
using Seaborn
using Distributions
```

「入門数理統計学」 p123  
定理２  
「$X$が平均$\mu$,分散$\sigma^2$の正規分布に従うとし，この正規母集団から大きさnの無作為標本がとられたとき．標本平均は$\overline{X}$は平均$\mu$，分散$\frac{\sigma^2}{n}の正規分布に従う$」

```julia
# 母集団とする．(μ=3,σ=1)
X = rand(Normal(3, 1), 100000)
X2 = rand(Normal(3, 1/sqrt(10)), 100000)
```

# 本来だとこうなってほしい

```julia
kdeplot(X,label="σ=1")
kdeplot(X2,label="σ=1√10")
legend()
```

```julia
# 母集団から抽出した標本平均
W = rand(X,10)
```

乱数の組み合わせ（無作為抽出なので）良い組み合わせが選ばれると，p123の図のようになる．（下図）

```julia
kdeplot(X, label="母集団")
kdeplot(W,label="標本")
legend()
```

## 失敗したケース

```julia
# 母集団から抽出した標本平均
W = rand(X,10)
kdeplot(X, label="母集団")
kdeplot(W,label="標本")
legend()
```
