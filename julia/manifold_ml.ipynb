```julia
ENV["PYTHON"] = "/Users/noriakioshita/.pyenv/shims/python"
using PyCall
Pkg.build("PyCall")
@pyimport sklearn.datasets.samples_generator as samples_generator
```

```julia
n_samples = 1500
noise = 0.05
# スイスロールのデータを呼び出し
X, _ = samples_generator.make_swiss_roll(n_samples, noise)
```

```julia
using Plots
```

```julia
plotly()
scatter(X[:,1],X[:,2],X[:,3], m=0)
```

```julia
@pyimport sklearn.cluster as cluster
```

```julia
# クラスタを3つでやる．
y_km = cluster.KMeans(n_clusters=3, init="k-means++", n_init=10, max_iter=300, tol=1e-04, random_state=0)[:fit_predict](X) +1
plotdata = [X [string(i) for i=y_km];]
```

```julia
x1_x = Float64[]
x1_y = Float64[]
x1_z = Float64[]

one = findin(plotdata[:,4],["1"])

for i in one
    for j in 1:3
        push!(x1_x,plotdata[i,1])
        push!(x1_y,plotdata[i,2])
        push!(x1_z,plotdata[i,3])
    end
end
```
