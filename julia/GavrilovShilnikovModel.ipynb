```julia
using DifferentialEquations
```

```julia
μ = 0.456
ϵ = 0.0357

function bluesky(du,u,p,t)
    du[1] = u[1]*(2+μ-10(u[1]^2+u[2]^2)) + u[3]^2 + u[2]^2+2u[2]
    du[2] = -u[3]^3-(1+u[2])*(u[3]^2+u[2]^2+2u[2])-4*u[1]+μ*u[2]
    du[3] = (1+u[2])*u[3]^2+u[1]^2-ϵ
end
```

```julia
u0 = [1.0;0.0;0.0]
tspan = (0.0,100.0)
prob = ODEProblem(bluesky,u0,tspan)
sol = solve(prob)
```

```julia
using Plots
plot(sol,vars=(1,2,3))
```
