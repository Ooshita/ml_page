```julia
using Plots

POPULATION = 20
GENETIC_LONG = 5
MAX_GENETIC_VALUE_LONG = 100
best_individual = 0

function roulette(individual_fitness_value, choised_roulette, elite)
    #ルー+レット選択法: 適合度に比例した割合で個体を選択する方法．
    # エリート選択で選ばれたものは除外
    individual_fitness_value[elite] = 0
    sum_fitness_value = 0

    # 適合度を合計する計算
    for i in range(POPULATION)
        sum_fitness_value += individual_fitness_value[i]
    end
    
    # 確率の計算 
    for i in range(POPULATION)
        # 確率＝適合度 / 適合の合計 * 100
        probability[i] = (individual_fitness_value[i] / sum_fitness_value) * 100
    end
    for i in range(POPULATION)
        if probability[i] <= random.randint(1, MAX_GENETIC_VALUE_LONG): 
            choised_roulette[0] = i
            break
    end
    for i in range(POPULATION)
        if probability[i] <= random.randint(1, MAX_GENETIC_VALUE_LONG)and i != choised_roulette[0]:
            choised_roulette[1] = i
            break
    end
end

print("Generation = ", end="")
generation_num = int(input())
current_generation = zeros((POPULATION, GENETIC_LONG))
individual_fitness_value = zeros([POPULATION])
probability_mutation = zeros([POPULATION])
choised_roulette = zeros([POPULATION])
elite = 0
next_generation = np.zeros((POPULATION, GENETIC_LONG))
probability = np.zeros([POPULATION])

print("") # 改行
for i in range(POPULATION):
print("POPULATION_" + str(i+1) + "= " ,end="")
for j in range(GENETIC_LONG):
    ran_num = random.randint(1, MAX_GENETIC_VALUE_LONG)
    current_generation[i][j] = ran_num
    print(str(int(current_generation[i][j])), end=" ")
print("") 
print("")

for h in range(generation_num):
for i in range(POPULATION):
    individual_list = []
    for j in range(GENETIC_LONG):
        # 適合度 = 世代の値の合計
        individual_fitness_value[i] += current_generation[i][j]
        individual_list.append(current_generation[i][j])
        # print("現在の世代の値=  " + str(current_generation[i][j]))
    # エリート選択法 
    best_individual += max(individual_list)
# print("エリート選択法" + str(best_individual))
for i in range(1, POPULATION): 
    print("fitness: " + str(individual_fitness_value[i]) + "  best: " + str(best_individual))
    if individual_fitness_value[i] > best_individual:
        best_individual = individual_fitness_value[i]
        elite = i # 最優秀の個体の番号を得る
        # print("elite= " + str(elite))
for j in range(GENETIC_LONG):
    # 最優秀の個体は次世代にそのままコピー
    next_generation[0][j] = current_generation[elite][j]
print("Generation = " + str(h+1) + ",elite = ", end="")
for j in range(GENETIC_LONG):
    print(str(int(current_generation[elite][j])), end=" ")
print("")

for i in range(POPULATION):
    probability_mutation[0] = random.randint(1, GENETIC_LONG)
    if probability_mutation[0] == 0.2: # 0.5%の確率で突然範囲
        roulette(individual_fitness_value, choised_roulette, elite) # ルーレット選択法
        current_generation[int(choised_roulette[0])][random.randint(0, GENETIC_LONG -1)]\
            = random.randint(1, GENETIC_LONG)
        # 適当な遺伝子の値が適当に書き換わる
        for j in range(GENETIC_LONG):
            next_generation[i][j] = current_generation[i][j]
    else: # 99%の確率で２点交叉
        roulette(individual_fitness_value, choised_roulette ,elite)
        start_cross = random.randint(1, GENETIC_LONG)
        end_cross = random.randint(1, GENETIC_LONG)
        while start_cross >= end_cross:
            for j in range(GENETIC_LONG):
                next_generation[i][j] = current_generation[int(choised_roulette[0])][j]
            j = start_cross
            for j in range(end_cross):
                next_generation[i][j] = current_generation[int(choised_roulette[1])][j]
            start_cross = random.randint(1, GENETIC_LONG)
            end_cross = random.randint(1, GENETIC_LONG)
for i in range(POPULATION):
    for j in range(GENETIC_LONG):
        current_generation[i][j] = next_generation[i][j]
best_individual = 0
individual_fitness_value = [0 for i in range(POPULATION)]
```
