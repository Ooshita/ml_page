```julia
using Distributions
```

```julia
d = MixtureModel(Gamma[
   Gamma(2.3),
   Gamma(5,3)], [0.2, 0.8])
x = rand(d,500000)
```

適当な混合分布のサンプル

```julia
histogram(x)
```

xは乱数で作った母集団ベクトルを入れる  
sizeには母集団から何個サンプリングするかの数  
repeatetionには何回繰り返すかの数

```julia
function center_limit(x, size, repetetion)
    means = Any[]
    anim = @animate for i in 1:repetetion
        samp = sample(x, size)
        push!(means, mean(samp))
        histogram(means)
    end
    gif(anim, "mygif2.gif", fps = 100)
    return means
end
```

```julia
result = center_limit(x, 50000, 5000)
```

```julia
display_img(mimetype, file) = open(file) do f
    base64text = base64encode(f)
    display("text/html", """<img src="data:$mimetype;base64,$base64text">""")
end

display_img("image/gif", "mygif2.gif")
```

```julia
histogram(result)
```

条件が成り立つ場合
