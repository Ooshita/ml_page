### k-means

```julia
function input_data()
    parse_data1= Float64[]
    parse_data2 = Float64[]
    open("./dataset/data_2.txt","r") do fp
        data = readlines(fp)
        for i in 1:length(data)
            push!(parse_data1,parse(Float64,split(data[i], " ")[1]))
            push!(parse_data2 ,parse(Float64,split(data[i], " ")[2]))
        end
    end
    X = [parse_data1 parse_data2]
    return X
end

X = input_data()

using Plots
gr()

function dist(μx, μy, xx, xy)
    return ((xx+xy)-(μx+μy))^2
end

# 重心の数
k = 5
# 重心の変数．
μx_old = Float64[]
μy_old = Float64[]
for i in 1:k
    # データの端を乱数の範囲にする．
    # 初期の重心を決める．
    push!(μx_old,rand(minimum(X[:,1]):0.1:maximum(X[:,1])))
    push!(μy_old,rand(minimum(X[:,2]):0.1:maximum(X[:,2])))
end

min = 1000000.0
cluster = 0
center_of_gravity_x = Float64[]
center_of_gravity_y = Float64[]
clusters = Int64[]
n = length(X[:,1])
    
μx_new = Float64[]
μy_new = Float64[]

# 収束判定用の変数
error = 100.0
error_arr = Float64[]

# 誤差関数が収束すれば終わり．
while error != 0.0  
    clusters = Int64[]
    
    min = 1000000.0 
    for i in 1:n
        # クラスタ割り当て
        # iには全データが入る．
        for j in 1:k
            distance = dist(μx_old[j], μy_old[j], X[i,1],X[i,2])
            if min > distance
                min = distance
                cluster = j
            end
        end
        push!(clusters, cluster)
    end
    
    plotdata = [X [string(i) for i=clusters];]
    
    # 主に重心の計算
    for i in 1:k
        cluster_len = length(findin(plotdata[:,3],["$i"]))
        for j in findin(plotdata[:,3],["$i"])
            # 下二行が重心の計算
            push!(center_of_gravity_x,((X[j,1]) / cluster_len))
            push!(center_of_gravity_y,((X[j,2]) / cluster_len))
        end
        push!(μx_new, sum(center_of_gravity_x))
        push!(μy_new, sum(center_of_gravity_y))
        center_of_gravity_x = Float64[]
        center_of_gravity_y = Float64[]
        error = dist(μx_new[i],μy_new[i],μx_old[i],μy_old[i])
        push!(error_arr, error)
    end
    μx_old = μx_new
    μy_old = μy_new
    μx_new = Float64[]
    μy_new = Float64[]
end

tmp_x = Float64[]
tmp_y = Float64[]
x_arr = []
y_arr = []

function result(cluster_num)
    tmp_x = Float64[]
    tmp_y = Float64[]
    for p in 1:cluster_num
        result = findin(plotdata[:,3],["$p"])
        for i in result
            push!(tmp_x, plotdata[i,1])
            push!(tmp_y, plotdata[i,2])
        end
        push!(x_arr, tmp_x)
        push!(y_arr, tmp_y)
        tmp_x = Float64[]
        tmp_y = Float64[]
    end
end

plotdata = [X [string(i) for i=clusters];]

result(k)
plotdata
```

```julia
scatter(x_arr, y_arr)
scatter!(μx_old, μy_old, label="center")
```

### k-means++

```julia
function input_data()
    parse_data1= Float64[]
    parse_data2 = Float64[]
    open("./dataset/data_2.txt","r") do fp
        data = readlines(fp)
        for i in 1:length(data)
            push!(parse_data1,parse(Float64,split(data[i], " ")[1]))
            push!(parse_data2 ,parse(Float64,split(data[i], " ")[2]))
        end
    end
    X = [parse_data1 parse_data2]
    return X
end

X = input_data()
n = length(X[:,1])

using Plots
gr()

function dist(μx, μy, xx, xy)
    return sqrt(((xx+xy)^2-(μx+μy)^2)^2)
end

# 重心の数
k = 2
# 重心の変数．
μx_old = Float64[]
μy_old = Float64[]

# データ点からランダムに選びそれを重心とする．
z1 = rand(1:1:length(X[:,1]))
push!(μx_old, X[z1,1])
push!(μy_old, X[z1,2])

DX = Float64[]
function centroid()
    DX = Float64[]
    max = 0
    max_num = 0
    for i in 1:n
        Dx = dist(X[i,1], X[i,2], X[z1,1], X[z1,2])
        push!(DX, Dx)
    end
    
    for i in 2:k
        for j in 1:n
            Dx = dist(X[j,1], X[j,2], X[z1,1], X[z1,2])
            Dx / sum(DX)
            if max < Dx
                max = Dx^2 / sum(DX)^2
                max_num = j
            end
        end
        push!(μx_old, X[max_num,1])
        push!(μy_old, X[max_num,2])
    end
end

centroid()


min = 1000000.0
cluster = 0
center_of_gravity_x = Float64[]
center_of_gravity_y = Float64[]
clusters = Int64[]
    
μx_new = Float64[]
μy_new = Float64[]

# 収束判定用の変数
error = 100.0
error_arr = Float64[]

# 誤差関数が収束すれば終わり．
while error != 0.0  
    clusters = Int64[]
    
    min = 1000000.0 
    for i in 1:n
        # クラスタ割り当て
        # iには全データが入る．
        for j in 1:k
            distance = dist(μx_old[j], μy_old[j], X[i,1],X[i,2])
            if min > distance
                min = distance
                cluster = j
            end
        end
        push!(clusters, cluster)
    end
    
    plotdata = [X [string(i) for i=clusters];]
    
    # クラスタ中心の計算
    for i in 1:k
        cluster_len = length(findin(plotdata[:,3],["$i"]))
        for j in findin(plotdata[:,3],["$i"])
            # 重心の計算
            push!(center_of_gravity_x,((X[j,1]) / cluster_len))
            push!(center_of_gravity_y,((X[j,2]) / cluster_len))
        end
        push!(μx_new, sum(center_of_gravity_x))
        push!(μy_new, sum(center_of_gravity_y))
        center_of_gravity_x = Float64[]
        center_of_gravity_y = Float64[]
        error = dist(μx_new[i],μy_new[i],μx_old[i],μy_old[i])
        push!(error_arr, error)
    end
    μx_old = μx_new
    μy_old = μy_new
    μx_new = Float64[]
    μy_new = Float64[]
end
plotdata = [X [string(i) for i=clusters];]

tmp_x = Float64[]
tmp_y = Float64[]
x_arr = []
y_arr = []

function result(cluster_num)
    tmp_x = Float64[]
    tmp_y = Float64[]
    for p in 1:cluster_num
        result = findin(plotdata[:,3],["$p"])
        for i in result
            push!(tmp_x, plotdata[i,1])
            push!(tmp_y, plotdata[i,2])
        end
        push!(x_arr, tmp_x)
        push!(y_arr, tmp_y)
        tmp_x = Float64[]
        tmp_y = Float64[]
    end
end

result(k)
plotdata
```
